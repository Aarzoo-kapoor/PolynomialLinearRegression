IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORTING DATASET

```python
df = pd.read_csv('Ice_cream selling data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

USING LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

VISUALISATION OF LINEAR REGRESSION MODEL

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('units of ice cream sold VS temperature ')
plt.xlabel('Temperature')
plt.ylabel('Units of ice cream sold')
plt.show()
```

PREDICTING USING THE LINEAR REAGRESSION MODEL

```python
lin_reg.predict([[3.6]])
"""as we can see the temperature at 3.610 the number of ice cream sold is
around 29 so at temperature 3.6 it must be around that number but my using
linear regression model its predicting 13 units of ice cream sold which is
not accurately correct plus from the graph we can see the model is not a
perfect fit model, so by polynomial regression model we can make a perfect
fit model"""
```

USING POLYNOMIAL REGRESSION MODEL

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

VISUALISATION OF POLYNOMIAL LINEAR REGRESSION MODEL

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('units of ice cream sold VS temperature ')
plt.xlabel('Temperature')
plt.ylabel('Units of ice cream sold')
plt.show()
```

VISUALISATION OF POLYNOMIAL LINEAR REGRESSSION MODEL WITH HIGHER VISUALISATION

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('units of ice cream sold VS temperature ')
plt.xlabel('Temperature')
plt.ylabel('Units of ice cream sold')
plt.show()
```

PREDICTING THROUGH POLYNOMAIL LINEAR REGRESSION

```python
lin_reg_2.predict(poly_reg.fit_transform([[3.6]]))
```

As we can see emperature at 3.610 the number of ice cream sold is around 29 so at temperature 3.6 it must be around that number and by using polynomail linear regression model we can see at temperature 3.6 the number of ice cream sold is 24 which is very much accurate and by our graph we can deduce out that polynomial linear regression model is the perfect fit for the dataset.

 Use polynomial regression when you see a curved relationship in your data that linear regression fails to capture, but be mindful of overfitting and complexity.**bold text**
